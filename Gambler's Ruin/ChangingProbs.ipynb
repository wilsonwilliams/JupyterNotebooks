```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
init_wealth = 1_000
init_p = 0.52
init_q = 1 - init_p
num_sims = 1_000
avg_gain = 5
avg_loss = 5
```

```python
wealth = np.empty(num_sims)
wealth[0] = init_wealth
```

```python
p = init_p
q = init_q
p_probs = np.empty(num_sims)
q_probs = np.empty(num_sims)
p_probs[0] = p
q_probs[0] = q
```

```python
for i in range(1, len(wealth)):
    rand = np.random.rand()
    p_probs[i] = p
    q_probs[i] = q
    if rand < p:
        wealth[i] = wealth[i - 1] + avg_gain
    else:
        wealth[i] = wealth[i - 1] - avg_loss
    prob_update = np.random.randn() * 0.007
    p += prob_update
    q -= prob_update
```

```python
fig, ax1 = plt.subplots()

ax1.plot(wealth, color='b', label="Wealth", alpha=0.5)
ax1.set_ylabel("Wealth ($)")

ax2 = ax1.twinx()
ax2.plot(p_probs, color='g', label='p', alpha=0.5)
ax2.plot(q_probs, color='r', label='q', alpha=0.5)
ax2.set_ylabel("Probability")

ax1.set_xlabel("Trial")

plt.title("Total Wealth With Varying Probability")
plt.tight_layout()
plt.show()
```
