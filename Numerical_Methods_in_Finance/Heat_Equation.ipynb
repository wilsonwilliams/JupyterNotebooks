```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
L = 10.0
T = 2.0
Nx = 100
Nt = 500
```

```python
dx = L / (Nx - 1)
dt = T / Nt
lambda_ = dt / dx**2
lambda_
```

```python
u = np.zeros(Nx)
u[int(Nx/3)] = 100
```

```python
for n in range(1, Nt + 1):
    u_new = np.copy(u)
    for j in range(1, Nx - 1):
        u_new[j] = u[j] + lambda_ * (u[j+1] - 2*u[j] + u[j-1])
    
    u = np.copy(u_new)

    if n % 50 == 0:
        plt.plot(np.linspace(0, L, Nx), u, label=f't={n*dt:.2f}s')

plt.xlabel('Position (x)')
plt.ylabel('Temperature (u)')
plt.title('Explicit Method with Heat Equation')
plt.legend()
plt.tight_layout()
plt.savefig('figures/heat_eq.jpg', dpi=300)
plt.show()
```
