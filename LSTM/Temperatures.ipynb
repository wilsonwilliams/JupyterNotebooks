```python
import tensorflow as tf
import os
import pandas as pd
import numpy as np
```

```python
zip_path = tf.keras.utils.get_file(
    origin='https://storage.googleapis.com/tensorflow/tf-keras-datasets/jena_climate_2009_2016.csv.zip',
    fname='jena_climate_2009_2016.csv.zip',
    extract=True)
csv_path, _ = os.path.splitext(zip_path)
```

```python
df = pd.read_csv(csv_path)
df
```

```python
df = df[5::6]
df
```

```python
df.index = pd.to_datetime(df["Date Time"], format="%d.%m.%Y %H:%M:%S")
df[:26]
```

```python
temp = df["T (degC)"]
type(temp), temp.plot()
```

```python
def df_to_X_y(df, window_size=5):
    df_as_np = df.to_numpy()
    X = []
    y = []
    for i in range(len(df_as_np) - window_size):
        row = [[a] for a in df_as_np[i : i + window_size]]
        X.append(row)
        label = df_as_np[i + window_size]
        y.append(label)
    return np.array(X), np.array(y)
```

```python
WINDOW_SIZE = 5
X, y = df_to_X_y(temp, WINDOW_SIZE)
X.shape, y.shape
```

```python
X_train, y_train = X[:60_000], y[:60_000]
X_val, y_val = X[60_000:65_000], y[60_000:65_000]
X_test, y_test = X[65_000:], y[65_000:]
X_train.shape, y_train.shape, X_val.shape, y_val.shape, X_test.shape, y_test.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import *
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras.metrics import RootMeanSquaredError
from tensorflow.keras.optimizers.legacy import Adam

model = Sequential()
model.add(InputLayer((5, 1)))
model.add(LSTM(64))
model.add(Dense(8, 'relu'))
model.add(Dense(1, 'linear'))

model.summary()
```

```python
cp = ModelCheckpoint('model/', save_best_only=True)
model.compile(loss=MeanSquaredError(), optimizer=Adam(learning_rate=1e-4), metrics=[RootMeanSquaredError()])
```

```python
model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=10, callbacks=[cp])
```

```python
from tensorflow.keras.models import load_model
model = load_model('model/')
```

```python
train_predictions = model.predict(X_train).flatten()
train_results = pd.DataFrame(data={'Train Predictions': train_predictions, 'Actual': y_train})
train_results
```

```python
import matplotlib.pyplot as plt

num_points_to_graph = 100
```

```python
plt.plot(train_results['Train Predictions'][:num_points_to_graph], color='g', label='Predictions')
plt.plot(train_results['Actual'][:num_points_to_graph], color='b', label='Actual')
plt.title('Training Predictions vs Actual')
plt.xlabel('Date')
plt.ylabel('Temp (degC)')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()
```

```python
val_predictions = model.predict(X_val).flatten()
val_results = pd.DataFrame(data={'Val Predictions': val_predictions, 'Actual': y_val})
val_results
```

```python
plt.plot(val_results['Val Predictions'][:num_points_to_graph], color='g', label='Predictions')
plt.plot(val_results['Actual'][:num_points_to_graph], color='b', label='Actual')
plt.title('Validation Predictions vs Actual')
plt.xlabel('Date')
plt.ylabel('Temp (degC)')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()
```

```python
test_predictions = model.predict(X_test).flatten()
test_results = pd.DataFrame(data={'Test Predictions': test_predictions, 'Actual': y_test})
test_results
```

```python
plt.plot(test_results['Test Predictions'][:num_points_to_graph], color='g', label='Predictions')
plt.plot(test_results['Actual'][:num_points_to_graph], color='b', label='Actual')
plt.title('Test Predictions vs Actual')
plt.xlabel('Date')
plt.ylabel('Temp (degC)')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()
```
